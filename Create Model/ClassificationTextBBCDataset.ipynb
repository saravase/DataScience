```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('datasets/bbc-text.csv')
data.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
x_train_count_vect = count_vectorizer.fit_transform(data['text'])

count_vectorizer.vocabulary_
```

```python
x_train_count_vect.shape
```

```python
print(x_train_count_vect[13])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer()
x_train_tfidf_vect = tfidf_vect.fit_transform(data['text'])

x_train_tfidf_vect.__dict__.keys()
```

```python
print(x_train_tfidf_vect[13])
```

```python
from sklearn.model_selection import train_test_split

y = data['category']
x_train, x_test, y_train, y_test = train_test_split(x_train_tfidf_vect, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(max_depth=10)
dtc.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = dtc.predict(x_test)

print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

```python
df_y = pd.DataFrame({
    'y_test' : y_test,
    'y_pred' : y_pred
})
df_y.head(10)
```
